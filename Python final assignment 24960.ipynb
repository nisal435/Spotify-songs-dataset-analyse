```python
# Import necessary libraries: pandas for data manipulation, matplotlib.pyplot for basic plotting,
# and seaborn for advanced statistical visualizations and graphics.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Spotify dataset
spotify_data = pd.read_csv("/Users/nisalgunawardhana/Desktop/spotify_songs.csv")
```

```python
# Data Collection and Loading
# Check the first few rows of the dataset to understand its structure
print(spotify_data.head())
```

```python
# Data Cleaning and Preprocessing
# Check for missing values
print(spotify_data.isnull().sum())
```

```python
# Handle missing values if any
spotify_data.dropna(inplace=True)
```

```python
# Visualization 1: Histogram of Track Popularity
plt.figure(figsize=(10, 6))
sns.histplot(spotify_data['track_popularity'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Track Popularity')
plt.xlabel('Track Popularity')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Boxplot of Energy by Playlist Genre
plt.figure(figsize=(12, 6))
sns.boxplot(x='playlist_genre', y='energy', data=spotify_data, palette='Set2')
plt.title('Energy Distribution by Playlist Genre')
plt.xlabel('Playlist Genre')
plt.ylabel('Energy')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 3: Scatter Plot of Danceability vs. Valence
plt.figure(figsize=(8, 6))
sns.scatterplot(x='danceability', y='valence', data=spotify_data, hue='playlist_genre', palette='Set2')
plt.title('Danceability vs. Valence')
plt.xlabel('Danceability')
plt.ylabel('Valence')
plt.show()
```

```python
# Visualization 4: Bar Plot of Mean Popularity by Playlist Genre
mean_popularity_by_genre = spotify_data.groupby('playlist_genre')['track_popularity'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='playlist_genre', y='track_popularity', data=mean_popularity_by_genre, palette='Set2')
plt.title('Mean Popularity by Playlist Genre')
plt.xlabel('Playlist Genre')
plt.ylabel('Mean Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 5: Violin Plot of Acousticness by Playlist Subgenre
plt.figure(figsize=(12, 6))
sns.violinplot(x='playlist_subgenre', y='acousticness', data=spotify_data, palette='Set3')
plt.title('Acousticness Distribution by Playlist Subgenre')
plt.xlabel('Playlist Subgenre')
plt.ylabel('Acousticness')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 6: Distribution of Speechiness by Playlist Genre
plt.figure(figsize=(10, 6))
sns.boxplot(x='playlist_genre', y='speechiness', data=spotify_data, palette='Set2')
plt.title('Speechiness Distribution by Playlist Genre')
plt.xlabel('Playlist Genre')
plt.ylabel('Speechiness')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 7: Count Plot of Playlist Genres
plt.figure(figsize=(10, 6))
sns.countplot(y='playlist_genre', data=spotify_data, palette='Set2', order=spotify_data['playlist_genre'].value_counts().index)
plt.title('Count of Songs by Playlist Genre')
plt.xlabel('Count')
plt.ylabel('Playlist Genre')
plt.show()
```

```python
# Visualization 8: Distribution of Track Durations
plt.figure(figsize=(10, 6))
sns.histplot(spotify_data['duration_ms'] / (1000 * 60), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Track Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 9: Scatter Plot of Danceability vs. Tempo with Popularity Color Gradient
plt.figure(figsize=(10, 6))
plt.scatter(spotify_data['danceability'], spotify_data['tempo'], c=spotify_data['track_popularity'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='Popularity')
plt.title('Danceability vs. Tempo')
plt.xlabel('Danceability')
plt.ylabel('Tempo')
plt.show()
```

```python
# Visualization 10: Scatter Plot of Loudness vs. Valence with Popularity Size
plt.figure(figsize=(10, 6))
plt.scatter(spotify_data['loudness'], spotify_data['valence'], s=spotify_data['track_popularity'], alpha=0.5)
plt.title('Loudness vs. Valence')
plt.xlabel('Loudness')
plt.ylabel('Valence')
plt.show()
```

```python
# Correct Application of Descriptive Statistics

# Summary Statistics for Numerical Columns
summary_statistics = spotify_data.describe()

# Mean, Median, and Standard Deviation for Specific Columns
mean_energy = spotify_data['energy'].mean()
median_energy = spotify_data['energy'].median()
std_energy = spotify_data['energy'].std()

mean_popularity = spotify_data['track_popularity'].mean()
median_popularity = spotify_data['track_popularity'].median()
std_popularity = spotify_data['track_popularity'].std()

# Print Summary Statistics
print("Summary Statistics for Numerical Columns:")
print(summary_statistics)

print("\nMean Energy:", mean_energy)
print("Median Energy:", median_energy)
print("Standard Deviation of Energy:", std_energy)

print("\nMean Popularity:", mean_popularity)
print("Median Popularity:", median_popularity)
print("Standard Deviation of Popularity:", std_popularity)
```

```python
# Clarity and Depth of Insights and Observations

# Descriptive Statistics Observations
# Analyze mean, median, and standard deviation for 'energy' and 'track_popularity'
energy_mean = spotify_data['energy'].mean()
energy_median = spotify_data['energy'].median()
popularity_mean = spotify_data['track_popularity'].mean()
popularity_median = spotify_data['track_popularity'].median()

print("Energy Mean:", energy_mean)
print("Energy Median:", energy_median)
print("Popularity Mean:", popularity_mean)
print("Popularity Median:", popularity_median)
```
